```python
# Concatenate the balance sheet dataframes
combined_balance_sheet = pd.concat([df1_1, df2_1,df3_1,df4_1], ignore_index=True)
```

```python
# Concatenate the balance sheet dataframes
combined_income_sheet = pd.concat([df12_2,df34_2], ignore_index=True)
combined_cash_sheet=pd.concat([df12_3,df34_3],ignore_index=True)
```

```python
# Extract the common companies and years
common_companies_years = set(combined_balance_sheet[['symbol', 'calendarYear']].itertuples(index=False, name=None)).intersection(
    set(combined_income_sheet[['symbol', 'calendarYear']].itertuples(index=False, name=None)),
    set(combined_cash_sheet[['symbol', 'calendarYear']].itertuples(index=False, name=None))
)

# Convert the set to a DataFrame for easy filtering
common_df = pd.DataFrame(list(common_companies_years), columns=['symbol', 'calendarYear'])
```

```python
# Merge the common_df with each dataframe to filter the data
filtered_balance_sheet = pd.merge(common_df, combined_balance_sheet, on=['symbol', 'calendarYear'])
filtered_income_statement = pd.merge(common_df, combined_income_sheet, on=['symbol', 'calendarYear'])
filtered_cashflow_statement = pd.merge(common_df, combined_cash_sheet, on=['symbol', 'calendarYear'])
```

```python
# Merge the filtered dataframes column-wise
final_df = filtered_balance_sheet.merge(
    filtered_income_statement.drop(columns=['symbol', 'calendarYear']),
    left_index=True, right_index=True, how='inner'
).merge(
    filtered_cashflow_statement.drop(columns=['symbol', 'calendarYear']),
    left_index=True, right_index=True, how='inner'
)

# Adding back the 'Company' and 'Year' columns
final_df = pd.concat([filtered_balance_sheet[['symbol', 'calendarYear']], final_df], axis=1)
```

```python
import yfinance as yf
import pandas as pd

# List of companies (using ticker symbols)
unique_companies_list = final_df['symbol'].unique().tolist()

# Define the time period
start_date = '2019-01-01'
end_date = '2023-12-31'

# Initialize an empty dictionary to store data
historical_data = {}

# Iterate over companies and fetch data
for company in unique_companies_list:
    try:
        ticker = yf.Ticker(company)
        historical_data[company] = ticker.history(start=start_date, end=end_date)
    except Exception as e:
        print(f"Error fetching data for {company}: {e}")

# Create a DataFrame to store the end-of-year share prices
end_of_year_prices = []

for company, data in historical_data.items():
    if not data.empty:
        data['calendarYear'] = data.index.year
        data = data.groupby('calendarYear').last().reset_index()
        data['symbol'] = company
        end_of_year_prices.append(data[['symbol', 'calendarYear', 'Close']])
    else:
        print(f"No data available for {company}")

# Combine all data into a single DataFrame
if end_of_year_prices:
    end_of_year_prices_df = pd.concat(end_of_year_prices)
    end_of_year_prices_df.rename(columns={'Close': 'SharePrice'}, inplace=True)
    # Display the end-of-year share prices
    print(end_of_year_prices_df)
else:
    print("No end-of-year prices were fetched.")
```

```python
# Extract the list of companies present in end_of_year_prices_df
valid_companies = end_of_year_prices_df['symbol'].unique()

# Filter final_df to include only those companies
filtered_final_df = final_df[final_df['symbol'].isin(valid_companies)]

# Merge filtered_final_df with end_of_year_prices_df on 'symbol' and 'calendarYear'
merged_df = pd.merge(filtered_final_df, end_of_year_prices_df, on=['symbol', 'calendarYear'], how='inner')

# Display the merged DataFrame
print(merged_df)
```
